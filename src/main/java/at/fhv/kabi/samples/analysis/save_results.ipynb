```python
import pandas as pd
import numpy as np
```

```python
df_size = pd.read_excel('../results/fileSizes.xlsx', names=["Use Cases", "Data Formats", "File Size"]).convert_dtypes()
df_speed = pd.read_excel('../results/elapsedTimes.xlsx', names=["Use Cases", "Data Formats", "Speed"], dtype={"Use Cases": str, "Data Formats": str, "Speed": np.float64})
```

```python
df_speed["Speed"] = df_speed["Speed"] / 1000 # convert to micro seconds
df_speed["Speed"] = round(df_speed["Speed"], 2)
```

```python
df_size = df_size.replace({'Data Formats': ['Cap\'n Proto (packed)', 'Cap\'n Proto (unpacked)', 'CSV (with header)', 'Ion Binary', 'Java Serialization']}, 
                {'Data Formats': ['Cap\'n (packed)', 'Cap\'n (unpacked)', 'CSV (header)', 'Ion (binary)', 'Java']})


df_speed = df_speed.replace({'Data Formats': ['Cap\'n Proto (packed)', 'Cap\'n Proto (unpacked)', 'CSV (with header)', 'Ion Binary', 'Java Serialization']}, 
                {'Data Formats': ['Cap\'n (packed)', 'Cap\'n (unpacked)', 'CSV (header)', 'Ion (binary)', 'Java']})
```

```python
df1 = df_size.pivot(index='Data Formats', columns='Use Cases', values='File Size')
df1.style.to_latex("./tables/absolute_sizes.tex")

df2 = df_speed.pivot(index='Data Formats', columns='Use Cases', values='Speed')
df2.style.format(precision=2).to_latex("./tables/absolute_speed.tex")
```

```python
dfd=pd.DataFrame()

for t in df1.keys():
    dft=pd.DataFrame()
    dft['Use Cases'] = list([t,t,t])
    w = list(df1.nsmallest(3,t)[t].index)
    dft['Data Formats'] = w
    x = list(df1.nsmallest(3,t)[t])
    dft['File Size']= x
    dfd = pd.concat([dfd,dft])

dfd.style.to_latex("./tables/top3_sizes.tex")
dfd

# HeartData als Index : 1, 2, 3 (untereinander)
```

```python
dfd2=pd.DataFrame()

for t in df2.keys():
    dft=pd.DataFrame()
    dft['Use Cases'] = list([t,t,t])
    w = list(df2.nsmallest(3,t)[t].index)
    dft['Data Formats'] = w
    x = list(df2.nsmallest(3,t)[t])
    dft['Serialization Speed']= x
    dfd2 = pd.concat([dfd2,dft])

dfd2.style.format(precision=2).to_latex("./tables/top3_speed.tex")
dfd2
```
